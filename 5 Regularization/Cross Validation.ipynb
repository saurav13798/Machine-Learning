# Separate X and Y :=>  Independent Feature  and Dependent feature 

```python
X = df.drop(columns=["id","Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

# Feature Engineering

## Data PreProcessing and data cleaning

```python
cat = list(X.columns[X.dtypes=="object"])
con = list(X.columns[X.dtypes!="object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="constant", fill_value="Not_Avl"),
    OneHotEncoder(handle_unknown='ignore',sparse_output=False)
)
```

```python
con_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
pre = ColumnTransformer(
    [
        ("cat",cat_pipe,cat),
        ("num",con_pipe,con)
    ]
).set_output(transform="pandas")
```

```python
pre
```

```python
xpre = pre.fit_transform(X)
xpre.head()
```

# apply Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xpre, Y, train_size=0.8, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

# Build a model

# Ridge Model

```python
from sklearn.linear_model import Ridge
model1 = Ridge(alpha=1)
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

## Hyper paramerter tuning

```python
import numpy as np
params = {
    "alpha" : np.arange(start=0.1, stop=100, step=0.1)
}
params
```

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
base_model = Ridge()
gscv = GridSearchCV(estimator=base_model,param_grid=params,scoring="r2")
gscv.fit(xtrain,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_ridge = gscv.best_estimator_
best_ridge
```

# Cross Validation of scores

```python
scores = cross_val_score(best_ridge,xtrain,ytrain,cv=5,scoring="r2") 
scores
```

```python
scores.mean()
```

```python
gscv2 = GridSearchCV(base_model, param_grid=params, scoring="r2", cv=3)
gscv2.fit(xtrain,ytrain)
```

```python
gscv2.score(xtrain,ytrain)
```

```python
gscv2.score(xtest,ytest)
```

```python
gscv2.best_params_
```

```python
best_ridge2 = gscv2.best_estimator_
best_ridge2
```

```python
scores2 = cross_val_score(best_ridge2,xtrain,ytrain,scoring="r2", cv=3)
scores2
```

```python
scores2.mean()
```
